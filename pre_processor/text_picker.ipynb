```python
import pandas as pd
```

```python
filename = "J20250829142736"
data = pd.read_csv(f"../data/{filename}_processed.csv")
data.head()
```

```python
products = pd.DataFrame(data["品名"].value_counts()).reset_index()
products.columns = ['product', 'count']
products.head()
```

```python
products_to_choose = products[products['count'] > 200]
products_to_choose.info()
```
